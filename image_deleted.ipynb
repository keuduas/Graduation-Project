```python
import os  
import random  

# 列出目录下所有图片文件  
def get_image_files(directory):  
    image_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp')  
    return [os.path.join(directory, f) for f in os.listdir(directory)   
            if os.path.isfile(os.path.join(directory, f)) and os.path.splitext(f)[1].lower() in image_extensions]  

# 主要逻辑  
def delete_images_to_keep_n(n):  
    image_files = get_image_files(dataset_dir)  
    if len(image_files) < n:  
        print(f"保留文件数{n}大于当前文件数{len(image_files)}，已保留所有文件。")  
        return  
    
    # 随机选择n张图片保留  
    keep_files = random.sample(image_files, n)  
    # 剩下的文件需要删除  
    delete_files = [f for f in image_files if f not in keep_files]  
    
    # 确认是否需要删除  
    # print(f"即将删除{len(delete_files)}张图片，保留{len(keep_files)}张。")  
    # confirm = input("确认吗？(yes/no): ")  
    # if confirm.lower() != "yes":  
    #     print("操作取消。")  
    #     return  
    
    # 执行删除操作  
    for file in delete_files:  
        try:  
            os.remove(file)  
            print(f"已删除文件：{file}")  
        except Exception as e:  
            print(f"删除文件{file}时发生错误：{str(e)}")  

```
